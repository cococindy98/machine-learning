```python
def Age_category(age):
    cat=''
    if age<=10 : cat=10
    elif age<=20 : cat=20
    elif age<=30 : cat=30
    elif age<=40 : cat=40
    elif age<=50 : cat=50
    elif age<=60 : cat=60
    elif age<=70 : cat=70
    else: cat=80
    return cat

titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x : Age_category(x))
```

```python
titanic_df
```

```python
titanic_df['Solo']=titanic_df['SibSp']+titanic_df['Parch']
titanic_df['Solo']=titanic_df['Solo'].apply(lambda x:  'No' if x==0 else 'Yes')
```

```python
titanic_df['Pclass']=titanic_df['Pclass'].apply(lambda x :'upper' if x==1 else('middle' if x==2 else 'lower')) 
```

```python
titanic_dummies=pd.get_dummies(titanic_df['Pclass'])
for i in ['Sex','Age_cat','Solo']:
    data=pd.get_dummies(titanic_df[i])
    titanic_dummies=pd.concat([titanic_dummies,data],axis=1)
titanic_dummies
```

```python
titanic_preprocessed=pd.get_dummies(titanic_df['Pclass'])
for i in [ 'Sex', 'Age_cat', 'Solo']:
    a=pd.get_dummies(titanic_df[i])
    titanic_preprocessed=pd.concat([titanic_preprocessed,a],axis=1)
    titanic_preprocessed['Survived']=titanic_df['Survived']   

titanic_preprocessed
```

```python
feature=titanic_preprocessed.drop('Survived',axis=1)
target=titanic_preprocessed['Survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(feature,target,test_size=0.2,random_state=0)
```

```python
dt_clf=DecisionTreeClassifier()
dt_clf.fit(X_train,y_train)
pred=dt_clf.predict(X_test)
accuracy_score(y_test,pred)
```
