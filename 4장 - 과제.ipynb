```python
cred.drop('Time',inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
def get_train_test_df(a):
    target=a.iloc[:,-1]
    feature=a.iloc[:,:-1]
    X_train,X_test,y_train,y_test=train_test_split(feature,target,test_size=0.3,stratify=target)
    return X_train,X_test,y_train,y_test
```

```python
X_train,X_test,y_train,y_test=get_train_test_df(cred)
```

```python
y_train.value_counts()/y_train.shape[0]
```

```python
y_test.value_counts()/y_test.shape[0]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,roc_auc_score
```

```python
lr_clf=LogisticRegression()
lr_clf.fit(X_train,y_train)
pred=lr_clf.predict(X_test)
pred_proba=lr_clf.predict_proba(X_test)[:,1]
def get_clf_eval(y_test,pred=None,pred_proba=None):
    con=confusion_matrix(y_test,pred)
    ac=accuracy_score(y_test,pred)
    re=recall_score(y_test,pred)
    pre=precision_score(y_test,pred)
    roc=roc_auc_score(y_test,pred_proba)
    print('행렬:{0},정확도:{1},재현율:{2},정밀도:{3},AUC:{4}'.format(con,ac,re,pre,roc))
```

```python
pred=lr_clf.predict(X_test)
pred_proba=lr_clf.predict_proba(X_test)[:,1]
get_clf_eval(y_test,pred,pred_proba)
```

```python
def get_model_train_eval(model,ftr_train,ftr_test,tgt_train,tgt_test):
    model.fit(ftr_train,tgt_train)
    pred=model.predict(ftr_test)
    pred_proba=model.predict_proba(ftr_test)[:,1]
    get_clf_eval(tgt_test,pred,pred_proba)
```

```python
def get_preprocessed_df(df):
    scaler=StandardScaler()
    amount_n=scaler.fit_transform(df['Amount'].values.reshape(-1,1))
    #변환된 Amount를 Amount_scaled로 피처명 변경 후 DataFrame 맨 앞으로 입력
    df.insert(0,'Amount_scaled',amount_n)
    
    df.drop(['Amount'],axis=1,inplace=True)
    return df
```

```python
def get_outlier(df,column,weight=1.5):
    #column에 해당하는 column만 추출, 1/4분위와 3/4분위지점을 np.precentile로 구함
    fraud=df[df['Class']==1][column]
    quantile_25=np.percentile(fraud.values,25)
    quantile_75=np.percentile(fraud.values,75)
    #IQR을 구하고 IQR에 1.5곱해 최댓값과 최솟값 지점 구함
    iqr=quantile_75-quantile_25
    iqr_weight=iqr*weight
    lowest_value=quantile_25-iqr_weight
    highest_value=quantile_75+iqr_weight
    
    outlier_index=fraud[(fraud<lowest_value)|(fraud>highest_value)].index
    return outlier_index



outlier_index=get_outlier(df=cred,column='V14',weight=1.5)
print('이상치 인덱스:',outlier_index)
```

```python
def get_preprocessed_df(df):
    amount_n=np.log1p(df['Amount'])
    df.insert(0,'Amount_scaled',amount_n)
    df.drop(['Amount','Time'],axis=1,inplace=True)
    
    outlier_index=get_outlier(df=cred,column='V14',weight=1.5)
    df.drop(outlier_index,acis=0,inplace=True)
```

```python
cred=pd.read_csv(r'C:\Users\PC\Desktop\새 폴더\creditcard.csv')

X_train,X_test,y_train,y_test=get_train_test_df(cred)

print("##로지스틱 회귀예측 성능##")
lr_clf=LogisticRegression()
get_model_train_eval(lr_clf,ftr_train=X_train,ftr_test=X_test,tgt_train=y_train,tgt_test=y_test)

print("##LightGBM 예측 성능##")
lgbm_clf=LGBMClassifier(n_estimators=100,num_leaves=64,n_jobs=-1,boost_from_average=False)
get_model_train_eval(lgbm_clf,ftr_train=X_train,ftr_test=X_test,tgt_train=y_train,tgt_test=y_test)
```
